```python
import os
import numpy as np
import matplotlib.pyplot as plt

IMAGES_PATH='.'

def save_fig(fig_id, tight_layout=False, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)

# Generate x values
x = np.linspace(-5, 5, 400)

# Calculate the hyperbolic tangent values for x
y = np.tanh(x)

# Create the plot
plt.figure(figsize=(5, 5))
plt.plot(x, y, label='$tanh(x)$', linewidth=4)
plt.xlabel('$x$')
plt.ylabel('$tanh(x)$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
save_fig('tanh', tight_layout=True)
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values
x = np.linspace(-10, 10, 4000)

# Calculate the hyperbolic tangent values for x
y = sigmoid(x)

# Create the plot
plt.figure(figsize=(5, 5))
plt.plot(x, y, label=r"$σ(x)=\frac{1}{1 + e^{-x}}$", linewidth=4)
plt.xlabel('$x$')
plt.ylabel('$σ(x)$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend(fontsize=15)
save_fig('sigmoid', tight_layout=True)
plt.show()
```
