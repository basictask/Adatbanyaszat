```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve

IMAGES_PATH = '.'
def save_fig(fig_id, tight_layout=False, fig_extension="png", resolution=500):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)

# Example ground truth and predicted scores (change these with your data)
ground_truth = [1, 0, 1, 1, 0, 0, 1, 0, 1, 1]
predicted_scores = [0.9, 0.7, 0.8, 0.6, 0.4, 0.3, 0.5, 0.2, 0.7, 0.6]

num_samples=50
ground_truth = np.random.randint(2, size=num_samples)
predicted_scores = np.random.rand(num_samples)

# Calculate precision and recall
precision, recall, _ = precision_recall_curve(ground_truth, predicted_scores)


# Plot the precision-recall curve
plt.figure(figsize=(5,5))
plt.plot(recall, precision, linewidth=4)
plt.fill_between(recall, precision, color='lightblue', alpha=0.8)
plt.xlabel('Visszahívás $(R)$', size=15)
plt.ylabel('Pontosság $(P(R))$', size=15)
plt.grid(True)
save_fig('od_12', tight_layout=False)
plt.show()
```
